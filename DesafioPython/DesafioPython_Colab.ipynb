```python
df2 = df.groupby('Department').size().sort_values(ascending=False)

cores = sns.color_palette("tab10", len(df2))

df2.plot(kind='bar', color=cores)
plt.xlabel('Departamento')
plt.ylabel('Quantidade de funcionários')
plt.title('Quantidade de funcionários por departamento')
plt.show()
```

Comando que gera um gráfico de barras exibindo a quantidade de funcionários por departamento, organizando os dados em ordem decrescente. Também foram adicionados título e legenda para melhorar a visualização.

Ao observar revela que o departamento de "Research & Development" concentra a maior parte dos funcionários da empresa.

utilizei a biblioteca seaborn para personalizar as cores da tabela e torna-las mais claras.


```python
education_counts = df.groupby('Education').size()

plt.plot(education_counts.index, education_counts.values, marker='o', linestyle='-')
plt.xlabel("Nivel de educação")
plt.ylabel("Quantidade de pessoas")
plt.title("Nivel de distribuição da educação")
plt.show()
```

Armazenei o agrupamento da coluna 'Education' em uma variável e, em seguida, plotei um gráfico de linhas com legendas nos eixos X e Y, além de adicionar um título para melhor compreensão.

Observa-se que a maioria dos funcionários da empresa possui o nível de educação 3, indicando que a maior parte da equipe tem uma formação de nível intermediário.

```python
import matplotlib.pyplot as plt
import seaborn as sns

atrito = df['Attrition'].value_counts()

cores = sns.color_palette("pastel", len(atrito))

plt.pie(atrito, labels=atrito.index, autopct='%1.1f%%', colors=cores)
plt.title("Atrito no Trabalho")
plt.axis("equal")
plt.show()
```

Armazenei o agrupamento da coluna 'Attrition' em uma variável e, em seguida, plotei um gráfico de pizza para visualizar a distribuição.

Pode-se observar que a maioria dos funcionários da empresa não apresenta atrito com outros integrantes, indicando um ambiente de trabalho relativamente estável.
